```python
import torch
from torch.nn.utils.rnn import pad_sequence

# Example list of sequences (tensors with different lengths)
sequence_a = torch.ones([1, 3, 3, 3])
sequence_b = torch.ones([4, 3, 3, 3])
sequence_c = torch.ones([
    6, 3, 3, 3])

list_of_sequences = [sequence_a, sequence_b, sequence_c]

# torch.stack(list_of_sequences, dim=0)  # This will raise an error due to different lengths
# Pad zero into the sequences to make them the same length
padded_sequences = pad_sequence(list_of_sequences, batch_first=True, padding_value=0)
print(padded_sequences[0])  # This will print the shape of the padded tensor
```

```python
import torch
from torch.distributions import Normal

# Assuming mean and logStd are tensors from a neural network
# Example shape: (batch_size, 1)
mean = torch.randn(7, 5, 1)
logStd = torch.randn(7, 5, 1)

# 1. Prepare parameters
mu = mean.squeeze(-1)       # Shape: (7, 5)
std = torch.exp(logStd).squeeze(-1) # Shape: (7, 5)

# 2. Define the distribution
dist = Normal(mu, std)

# 3. Sample (non-differentiable)
sample = dist.sample()
print(sample)  # Should be (7, 5)

# 4. Or, sample using reparameterization trick (differentiable)
sample_diff = dist.rsample()
print(sample_diff)  # Should also be (7, 5)
```
