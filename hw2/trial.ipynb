```python
%matplotlib inline
import matplotlib.pyplot as plt

# ----------------------
# Data
# ----------------------
horizon = [5, 10, 15, 20, 25]
success_rate = [0, 0, 0, 0, 0]
avg_reward = [0.887, 0.951, 0.971, 0.98, 0.954]

# ----------------------
# Create figure
# ----------------------
fig, ax1 = plt.subplots(figsize=(8, 5))

# ---- Success Rate (Left Axis) ----
ax1.plot(
    horizon,
    success_rate,
    marker='o',
    linewidth=2.5,
    label="Success Rate"
)

ax1.set_xlabel("Planning Horizon", fontsize=12)
ax1.set_ylabel("Success Rate", fontsize=12)
ax1.set_ylim(-0.05, 1.0)
ax1.grid(True, linestyle="--", alpha=0.5)

# ---- Reward (Right Axis) ----
ax2 = ax1.twinx()

ax2.plot(
    horizon,
    avg_reward,
    marker='s',
    linewidth=2.5,
    linestyle='--',
    label="Average Reward"
)

ax2.set_ylabel("Average Reward", fontsize=12)
ax2.set_ylim(0.85, 1.0)

# ----------------------
# Legend (combine both axes)
# ----------------------
lines_1, labels_1 = ax1.get_legend_handles_labels()
lines_2, labels_2 = ax2.get_legend_handles_labels()

ax1.legend(
    lines_1 + lines_2,
    labels_1 + labels_2,
    loc="upper left"
)

# ----------------------
# Title
# ----------------------
plt.title("Success Rate and Reward vs Planning Horizon", fontsize=14)

plt.tight_layout()

# Save figure
plt.savefig("horizon_performance.png", dpi=300)
plt.show()
```

```python
import torch
from torch.distributions import Normal

# Assuming mean and logStd are tensors from a neural network
# Example shape: (batch_size, 1)
mean = torch.randn(7, 5, 1)
logStd = torch.randn(7, 5, 1)

# 1. Prepare parameters
mu = mean.squeeze(-1)       # Shape: (7, 5)
std = torch.exp(logStd).squeeze(-1) # Shape: (7, 5)

# 2. Define the distribution
dist = Normal(mu, std)

# 3. Sample (non-differentiable)
sample = dist.sample()
print(sample)  # Should be (7, 5)

# 4. Or, sample using reparameterization trick (differentiable)
sample_diff = dist.rsample()
print(sample_diff)  # Should also be (7, 5)
```
